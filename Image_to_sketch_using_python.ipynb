**IMAGE TO PENCIL SKETCH CONVERSION USING PYTHON**

OpenCV (Open Source Computer Vision Library) is a powerful library for computer vision and image processing tasks. It provides various functions and tools to manipulate, process, and analyze images and videos.

```python
import cv2
```

```python
# Loading files from the colab
from google.colab import files
uploaded = files.upload()
```

**Loading the image**

```python
# Read an image
img = cv2.imread('puppy.jpeg', 1)
```

```python
import matplotlib.pyplot as plt

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

**Converting image into gray_scale image**

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray)
```

**Inverting the image**

We are using the bitwise_not function which is used to make brighter regions lighter and vice versa so that we can find the edges to create a pencil sketch.

```python
img_invert = cv2.bitwise_not(img_gray)

plt.imshow(img_invert)
```

**Smoothing the image**

We have used the gaussian blur technique with 21 x 21 pixel and the default sigma values filter on the image to smoothen our image. By increasing the filter size, we can create thin lines for our sketch and it is used to reduce the noise in the image.

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)

plt.imshow(img_smoothing)
```

**Final Sketch**

```python
inverted_blurred = 255 - img_smoothing
pencil_sketch = cv2.divide(img_gray, inverted_blurred, scale=256.0)
```

```python
plt.imshow(pencil_sketch, cmap='gray')  # Specify cmap='gray' for grayscale images
plt.axis('off')  # Turn off axis labels
plt.title('Pencil Sketch')
plt.show()
```

**Result**

```python
plt.figure(figsize=(10, 5))

# Plot the original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Original Image')
plt.axis('off')

# Plot the pencil sketch
plt.subplot(1, 2, 2)
plt.imshow(pencil_sketch, cmap='gray')
plt.title('Pencil Sketch')
plt.axis('off')

# Adjust layout
plt.tight_layout()

# Show the figure
plt.show()
```
